### 2. Search Functions

Now, let's implement the `get_faq_by_id` and `get_faqs_by_category` functions.

```python
def get_faq_by_id(faqs: list, faq_id: str) -> dict | None:
    """
    Returns the FAQ with the given id, or None if not found.

    Args:
        faqs (list): The list of FAQ dictionaries.
        faq_id (str): The unique identifier of the FAQ to retrieve.

    Returns:
        dict | None: The FAQ dictionary if found, otherwise None.
    """
    # Iterate through the list of FAQs
    for faq in faqs:
        # Check if the current FAQ's id matches the requested faq_id
        if faq.get("id") == faq_id:
            return faq
    # If no matching FAQ is found after checking all entries, return None
    return None

print("Function `get_faq_by_id` implemented.")
```

```python
def get_faqs_by_category(faqs: list, category: str) -> list:
    """
    Returns all FAQs belonging to the given category (case-insensitive).

    Args:
        faqs (list): The list of FAQ dictionaries.
        category (str): The category to filter FAQs by.

    Returns:
        list: A list of FAQ dictionaries belonging to the specified category.
    """
    matching_faqs = []
    # Convert the search category to lowercase for case-insensitive matching
    search_category_lower = category.lower()
    # Iterate through each FAQ in the knowledge base
    for faq in faqs:
        # Check if the FAQ's category (in lowercase) matches the search category
        if faq.get("category", "").lower() == search_category_lower:
            matching_faqs.append(faq)
    return matching_faqs

print("Function `get_faqs_by_category` implemented.")
```

```python
def _task1_search_by_keyword(faqs: list, query: str) -> list:
    """
    Searches FAQs by matching query words against keywords, question, and category fields.
    Matching is case-insensitive.
    Returns matching FAQs ordered by number of keyword hits (most matches first).

    Args:
        faqs (list): The list of FAQ dictionaries.
        query (str): The search query string.

    Returns:
        list: A list of matching FAQ dictionaries, ordered by the number of keyword hits.
    """
    query_words = [word.lower() for word in query.split()]
    scored_faqs = []

    for faq in faqs:
        hit_count = 0
        # Prepare FAQ fields for case-insensitive matching
        faq_question_lower = faq.get("question", "").lower()
        faq_category_lower = faq.get("category", "").lower()
        faq_keywords_lower = [kw.lower() for kw in faq.get("keywords", [])]

        for word in query_words:
            # Check for matches in question, category, and keywords
            if word in faq_question_lower:
                hit_count += 1
            if word in faq_category_lower:
                hit_count += 1
            if word in faq_keywords_lower:
                hit_count += 1

        if hit_count > 0:
            scored_faqs.append((hit_count, faq))

    # Sort by hit_count in descending order
    scored_faqs.sort(key=lambda x: x[0], reverse=True)

    # Return only the FAQ dictionaries
    return [faq for count, faq in scored_faqs]

print("Function `_task1_search_by_keyword` implemented.")
```

## Task 3: Intelligent FAQ Matching

This task focuses on improving FAQ matching beyond simple keyword searches by implementing TF-IDF vectorization and cosine similarity. We'll create an `FAQMatcher` class, a `hybrid_search` function, and demonstrate their effectiveness compared to the keyword search.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

print("Libraries for TF-IDF and cosine similarity imported.")
```

### 1. FAQMatcher Class

Now, let's implement the `FAQMatcher` class using TF-IDF vectorization and cosine similarity.

```python
class FAQMatcher:
    """
    A class to match user queries to FAQs using TF-IDF vectorization and cosine similarity.
    """
    def __init__(self, faqs: list):
        """
        Initializes the FAQMatcher by building a TF-IDF index over FAQ questions and keywords.

        Args:
            faqs (list): A list of FAQ dictionaries, each with 'question' and 'keywords' fields.
        """
        self.faqs = faqs
        # Prepare the corpus by combining question and keywords for each FAQ
        # Ensure keywords list is joined into a string for vectorizer
        self.corpus = [
            faq['question'] + ' ' + ' '.join(faq.get('keywords', []))
            for faq in faqs
        ]
        # Initialize TfidfVectorizer
        self.vectorizer = TfidfVectorizer()
        # Fit the vectorizer to the corpus and transform the corpus into TF-IDF vectors
        self.faq_vectors = self.vectorizer.fit_transform(self.corpus)
        print("FAQMatcher initialized and TF-IDF index built.")

    def match(self, query: str, top_k: int = 3) -> list[tuple]:
        """
        Vectorizes the user query and computes cosine similarity against all FAQs.

        Args:
            query (str): The user's search query.
            top_k (int): The number of top matches to return.

        Returns:
            list[tuple]: A list of (faq_dict, confidence_score) tuples, sorted by score.
                         Scores are rounded to 4 decimal places.
        """
        # Vectorize the query using the same vectorizer fitted to the FAQ corpus
        query_vector = self.vectorizer.transform([query])
        # Compute cosine similarity between the query vector and all FAQ vectors
        cosine_similarities = cosine_similarity(query_vector, self.faq_vectors).flatten()

        # Combine FAQs with their similarity scores
        scored_faqs = []
        for i, score in enumerate(cosine_similarities):
            scored_faqs.append((self.faqs[i], round(score, 4)))

        # Sort by confidence score in descending order
        scored_faqs.sort(key=lambda x: x[1], reverse=True)

        return scored_faqs[:top_k]

    def best_match(self, query: str, threshold: float = 0.15) -> tuple | None:
        """
        Returns the single best match if its score meets or exceeds the threshold.

        Args:
            query (str): The user's search query.
            threshold (float): The minimum confidence score required for a match.

        Returns:
            tuple | None: A (faq_dict, confidence_score) tuple if a match meets the threshold,
                          otherwise None.
        """
        # Get the top 1 match using the match method
        top_match = self.match(query, top_k=1)

        # Check if a match exists and its score meets the threshold
        if top_match and top_match[0][1] >= threshold:
            return top_match[0]
        return None

    def explain_match(self, query: str) -> str:
        """
        Returns a formatted string showing the top 3 matches with their scores.

        Args:
            query (str): The user's search query.

        Returns:
            str: A formatted string of the top matches.
        """
        top_matches = self.match(query, top_k=3)
        explanation = []
        for i, (faq, score) in enumerate(top_matches):
            explanation.append(f"  {i+1}. [{score:.4f}] {faq['question']}")
        return "\n".join(explanation)


# Initialize the FAQMatcher with the existing FAQs from Task 1
faq_matcher = FAQMatcher(faqs)
print("\nFAQMatcher class and instance created.")
```

### 2. Hybrid Search

Now let's implement the `hybrid_search` function that combines keyword search and TF-IDF matching.

```python
def hybrid_search(faqs: list, query: str, faq_matcher: FAQMatcher, top_k: int = 3) -> list[tuple]:
    """
    Combines keyword search and TF-IDF matching to provide a hybrid search result.

    Args:
        faqs (list): The list of FAQ dictionaries.
        query (str): The user's search query.
        faq_matcher (FAQMatcher): An instance of the FAQMatcher class.
        top_k (int): The number of top results to return.

    Returns:
        list[tuple]: A list of (faq_dict, combined_score) tuples, sorted by score descending.
    """
    results = {}

    # 1. Keyword search (from Task 1) - assign a base score of 0.5
    keyword_hits = _task1_search_by_keyword(faqs, query)
    for faq in keyword_hits:
        faq_id = faq['id']
        # Add a base score for keyword matches, ensure it's not overridden by lower scores
        results[faq_id] = {'faq': faq, 'score': max(results.get(faq_id, {'score': 0})['score'], 0.5)}

    # 2. TF-IDF matching (using FAQMatcher) - use cosine similarity scores
    tfidf_matches = faq_matcher.match(query, top_k=len(faqs)) # Get all potential matches
    for faq, score in tfidf_matches:
        faq_id = faq['id']
        # Take the maximum score if an FAQ is found by both methods
        results[faq_id] = {'faq': faq, 'score': max(results.get(faq_id, {'score': 0})['score'], score)}

    # Convert dictionary to a sorted list of tuples (faq_dict, confidence_score)
    sorted_results = sorted(
        [(item['faq'], item['score']) for item in results.values()],
        key=lambda x: x[1],
        reverse=True
    )

    return sorted_results[:top_k]

print("Function `hybrid_search` implemented.")
```

```python
import os
import requests
from google.colab import userdata

# Retrieve API key from Colab secrets
OPENROUTER_API_KEY = userdata.get('OPENROUTER_API_KEY')

# Define the model to be used
MODEL_NAME = "openai/gpt-4o-mini"
```

```python
class LLMClient:
    def __init__(self, api_key: str, model: str):
        if not api_key:
            raise ValueError("API key cannot be empty.")
        self.api_key = api_key
        self.model = model
        self.api_base = "https://openrouter.ai/api/v1"

    def generate(self, prompt: str, system_message: str = None, max_tokens: int = 512) -> str:
        headers = {
            "Authorization": f"Bearer {self.api_key}",
            "Content-Type": "application/json"
        }

        messages = []
        if system_message:
            messages.append({"role": "system", "content": system_message})
        messages.append({"role": "user", "content": prompt})

        payload = {
            "model": self.model,
            "messages": messages,
            "max_tokens": max_tokens
        }

        try:
            response = requests.post(f"{self.api_base}/chat/completions", headers=headers, json=payload)
            response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
            return response.json()['choices'][0]['message']['content']
        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")
            print(f"Response body: {response.text}")
            raise
        except requests.exceptions.ConnectionError as conn_err:
            print(f"Connection error occurred: {conn_err}")
            raise
        except requests.exceptions.Timeout as timeout_err:
            print(f"Timeout error occurred: {timeout_err}")
            raise
        except requests.exceptions.RequestException as req_err:
            print(f"An unexpected error occurred: {req_err}")
            raise
        except KeyError:
            print("Error parsing response from LLM API. Missing 'choices' or 'message' key.")
            print(f"Full response: {response.json()}")
            raise

    def generate_faq_response(self, user_question: str, faq_entry: dict) -> str:
        faq_question = faq_entry['question']
        faq_answer = faq_entry['answer']

        system_prompt = """You are SupportAI, a friendly and professional support agent.
Your task is to rephrase an official FAQ answer to respond to a user's question,
while strictly adhering to the following rules:
1. Answer ONLY from the provided FAQ content. Do not invent facts or information beyond what is stated.
2. Maintain a friendly, professional, and helpful tone.
3. Keep responses concise, ideally under 150 words.
4. If the provided FAQ content does not fully answer the user's question, state that clearly and offer further assistance (e.g., 'If this doesn't fully answer your question, please let us know!'). Do not try to guess or extrapolate."""

        user_prompt = (
            f"Official FAQ Question: {faq_question}\n"
            f"Official FAQ Answer: {faq_answer}\n\n"
            f"User's Question: {user_question}\n\n"
            "Please rephrase the official FAQ answer to directly address the user's question. "
            "Remember to follow all the rules in your system prompt."
        )

        return self.generate(user_prompt, system_message=system_prompt, max_tokens=150)

# Initialize the LLM client
llm_client = LLMClient(api_key=OPENROUTER_API_KEY, model=MODEL_NAME)
```
